```python
df = pd.read_csv(r'C:\Users\Pranjal Shah\ecommerce-analytics\data\superstore.csv', encoding='latin-1')

print(f"Dataset loaded: {df.shape[0]:,} rows × {df.shape[1]} columns")
df.head(3)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date']  = pd.to_datetime(df['Ship Date'])

print("Date types after conversion:")
print(df[['Order Date', 'Ship Date']].dtypes)
print(f"\nDate range: {df['Order Date'].min().date()} → {df['Order Date'].max().date()}")
print(f"Dataset spans {df['Order Date'].dt.year.nunique()} years of data")
```

```python
df['Profit Margin %'] = (df['Profit'] / df['Sales'] * 100).round(2)
df['Days to Ship']    = (df['Ship Date'] - df['Order Date']).dt.days
df['Year']            = df['Order Date'].dt.year
df['Month']           = df['Order Date'].dt.month
df['Month Name']      = df['Order Date'].dt.strftime('%b')
df['YearMonth']       = df['Order Date'].dt.to_period('M')

def discount_bucket(d):
    if d == 0:       return '0% — No Discount'
    elif d <= 0.10:  return '1–10%'
    elif d <= 0.20:  return '11–20%'
    elif d <= 0.30:  return '21–30%'
    else:            return '30%+'

df['Discount Bucket'] = df['Discount'].apply(discount_bucket)

print("Derived metrics added ✓")
print(f"New columns added: {['Profit Margin %', 'Days to Ship', 'Year', 'Month', 'YearMonth', 'Discount Bucket']}")
```

```python
text_cols = ['Segment', 'Category', 'Sub-Category', 'Region', 'Ship Mode', 'State', 'City']

for col in text_cols:
    df[col] = df[col].str.strip().str.title()

print("Text columns standardized ✓")
for col in ['Segment', 'Category', 'Region', 'Ship Mode']:
    print(f"\n{col}: {sorted(df[col].unique())}")
```

```python
print("=== BUSINESS SANITY CHECKS ===")

neg_sales = df[df['Sales'] <= 0]
print(f"Orders with zero/negative Sales: {len(neg_sales)}")

neg_profit = df[df['Profit'] < 0]
print(f"Loss-making order lines: {len(neg_profit):,} ({len(neg_profit)/len(df)*100:.1f}% of all orders)")
print(f"Total profit destroyed: ${neg_profit['Profit'].sum():,.0f}")

print(f"\nAverage days to ship: {df['Days to Ship'].mean():.1f} days")
print(f"Shipping time range: {df['Days to Ship'].min()}–{df['Days to Ship'].max()} days")
```

```python
df.to_csv(r'C:\Users\Pranjal Shah\ecommerce-analytics\data\superstore_clean.csv', index=False)

print("=== CLEANING COMPLETE ✓ ===")
print(f"Clean dataset saved")
print(f"Final shape: {df.shape[0]:,} rows × {df.shape[1]} columns")
```
